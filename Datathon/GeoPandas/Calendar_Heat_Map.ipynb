```python
import numpy as np; 
np.random.seed(sum(map(ord, 'calmap')))
import pandas as pd
import calmap
import datetime
```

```python
weather = pd.read_csv('/home/shade/DS4A/Data/Dataset/Dataset/weather.csv')
weather.head()

pd.to_datetime(weather['date'])
```

```python
All_days = pd.to_datetime(weather['date'])
days = All_days.to_numpy()
events = pd.Series(np.random.randn(len(days)), index=days)
#calmap.yearplot(events, year=2015)
array_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

calmap.calendarplot(events, monthticks=1, daylabels= array_days,
                    dayticks=[0, 1, 2, 3, 4, 5, 6], cmap='YlGn',
                    fillcolor='grey', linewidth=0,
                    fig_kws=dict(figsize=(12, 8)))
```
