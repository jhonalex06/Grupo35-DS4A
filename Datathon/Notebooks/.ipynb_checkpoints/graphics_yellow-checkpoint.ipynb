```python
import pandas as pd
import numpy as np
import geopandas as gpd
import folium
from folium.plugins import TimeSliderChoropleth
import branca
import contextily as ctx
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, MultiPoint
```

```python
ny_shp = gpd.read_file("/home/shade/DS4A/Data/Dataset/Shapes/nta_map.shp")
yellow_taxi = pd.read_csv('/home/shade/DS4A/Grupo35-DS4A/Datathon/Notebooks/yellow_taxi_filtered.csv')
```

```python
yellow_taxi_Points =[Point(x, y) for x, y in zip(yellow_taxi['pickup_longitude'], yellow_taxi['pickup_latitude'])]
geo_yellow_taxi = gpd.GeoDataFrame(yellow_taxi, geometry=yellow_taxi_Points)
```

```python
fig, ax = plt.subplots(figsize = (20,16))
ny_shp.plot(figsize=(20,16), alpha=0.1, edgecolor='k', ax=ax)
geo_yellow_taxi.plot(figsize=(20,16), edgecolor='green', color='green', ax=ax)
```

```python
yellow_taxi.head()
```

```python
count = []
ntacode = []
ntaname = []

for i in range(len(ny_shp['geometry'])):
    count.append(geo_yellow_taxi[geo_yellow_taxi.intersects(ny_shp.loc[i, 'geometry'])].shape[0])
    ntacode.append(ny_shp.loc[i, 'ntacode'])
    ntaname.append(ny_shp.loc[i, 'ntaname'])
    if i == 50 or i == 100 or i == 150 or i == 180:
        print (i)
```

```python
df = pd.DataFrame(list(zip(ntacode, ntaname, count)), columns =['ntacode', 'ntaname', 'count']) 
df.to_csv('count_yellow_taxi_by_nta.csv')
```

```python
df = pd.read_csv('count_yellow_taxi_by_nta.csv')
```

```python
plt.figure(figsize=(30,20))
plt.xticks(rotation=90)
plt.bar(df['ntacode'],df['count'])
plt.savefig('bar_plot', format="png")
```

```python
df_merge_col = pd.merge(ny_shp, df, left_on=['ntacode'], right_on=['ntacode'])
df_merge_col.head()
```

```python
min_cn, max_cn = df_merge_col['count'].quantile([0.01,0.99]).apply(round, 2)

colormap = branca.colormap.LinearColormap(
    colors=['white','yellow','orange','red','darkred'],
    #index=beat_cn['count'].quantile([0.2,0.4,0.6,0.8]),b
    vmin=min_cn,
    vmax=max_cn
)

colormap.caption="Total crimes in Chicago by police beats"
```

```python
# interactive visualization for beat-specific crime rate in 2017

m_crime = folium.Map(location=[40.738, -73.98],
                        zoom_start=11,
                        tiles="OpenStreetMap")

style_function = lambda x: {
    'fillColor': colormap(x['properties']['count']),
    'color': 'black',
    'weight':2,
    'fillOpacity':0.5
}

stategeo = folium.GeoJson(
    df_merge_col.to_json(),
    name='New York Transport',
    style_function=style_function,
    tooltip=folium.GeoJsonTooltip(
        fields=['ntacode', 'count'],
        aliases=['ntacode', 'Total yellow Taxi'], 
        localize=True
    )
).add_to(m_crime)

colormap.add_to(m_crime)
m_crime.save('index.html')
```
