```python
for i in geojson['features']:
    print (i['properties'])
```

```python
features = []
geo = {}
lon = []
lat = []

for i in geojson['features']:
    if i['properties']['CODIGO_LOC'] == '1':
        features.append(i)
        lon.append(i['geometry']['coordinates'][0][0][0][0])
        lat.append(i['geometry']['coordinates'][0][0][0][1])
    print (i['properties']['NOMBRE'])

geo['features'] = features
```

```python
max(lon)
```

```python
min(lon)
```

```python
min(lon) + (max(lon) - min(lon))/2
```

```python
min(lat) + (max(lat) - min(lat))/2
```
