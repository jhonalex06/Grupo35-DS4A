```python
# Libraries needed for basic web-scraping
from IPython.core.display import HTML
from bs4 import BeautifulSoup
from IPython.display import IFrame
import urllib # package required to interact with live webpage
import requests
import pandas as pd
import ast# will use to store the data from the webpage
import csv
```

```python
site_url = 'http://iboca.ambientebogota.gov.co/mapa'
r = requests.get(site_url, 'html.parser')
soup = BeautifulSoup(r.content)
soup
```

```python
int(soup.find_all('span',{'id':'lblNumInm'})[0].text.split()[0].replace(',',''))
```

```python
path = []
links = soup.find_all('li',{"class":"title-grid"})
for i in links:
    path.append(i.find('a').get('href'))
    
len(path)
```

```python
site_url = 'https://www.fincaraiz.com.co{}'.format(path[1])
r = requests.get(site_url, 'html.parser')
soup = BeautifulSoup(r.content)
len(soup.find_all('script'))
```

```python
for i in soup.find_all('script',{"type":"text/javascript"}):
    if str(i).find('var sfAdvert') > 0:
        new_soup = i
```

```python
Data = ast.literal_eval(new_soup.text.split('var sfAdvert = ')[1].split(';\r\n')[0])
Data
```

```python
print(Data['ClientName'])
```

```python
csvfile = "example.csv"
with open(csvfile, "a") as output:
    writer = csv.writer(output, lineterminator='\n')
    A1 = Data['ClientName']
    A2 = Data['Longitude']
    writer.writerow([A1, A2])
```
