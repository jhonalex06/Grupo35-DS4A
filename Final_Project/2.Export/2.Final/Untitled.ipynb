```python
import pandas as pd
import geopandas as gpd
import re
```

```python
df = pd.read_csv('OLX_Properati.csv', sep=';')
df.shape
```

```python
df = df[df['tipo_inmueble'].notnull()]
```

```python
tipo_inmueble = []
for i in df['tipo_inmueble']:
    if 'casa' in i.lower():
        if 'apartamento' in i.lower():
            tipo_inmueble.append('apartamento y casa')
        else:
            tipo_inmueble.append('casa')
    else:
        if 'apartamento' in i.lower():
            tipo_inmueble.append('apartamento')
        else:
            tipo_inmueble.append('no_data')
```

```python
df['tipo_inmueble'] = tipo_inmueble
```
